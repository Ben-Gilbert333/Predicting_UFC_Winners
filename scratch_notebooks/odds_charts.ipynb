```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the range of American odds
american_odds = range(100, 1001, 50)

# Initialize an empty list to store the results
results = []

# Calculate the outcome for each American odds value
for odds in american_odds:
    # Calculate the payout for a $100 bet
    payout = 100 * (odds / 100)
    
    
    # Calculate the total result (profit or loss)
    results.append(payout)

# Create a line chart
plt.figure(dpi=300)
plt.plot(american_odds, results, color='limegreen')
plt.title('Betting $100 on Plus Odds')
plt.xlabel('American Odds')
plt.ylabel('Profit ($)')
plt.grid(False)

# Save the chart
plt.savefig('plus_odds.png')
# Show the chart
plt.show()





```

```python
# Define the range of American odds
american_odds = range(-1000, -109, 50)

# Initialize an empty list to store the results
results = []

# Calculate the outcome for each American odds value
for odds in american_odds:
    # Calculate the payout for a $100 bet
    payout = 100 / (odds / -100)
    
    # Calculate the total result (profit or loss)
    results.append(payout)

# Create a line chart
plt.figure(dpi=300)
plt.plot(american_odds, results, color='red')
plt.title('Betting $100 on Minus Odds')
plt.xlabel('American Odds')
plt.ylabel('Profit ($)')
plt.grid(False)

# Save the chart
plt.savefig('minus_odds.png')

# Show the chart
plt.show()

```
